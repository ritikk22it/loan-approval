```python
# define numerical & categorical columns
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
# Name of the columns
df.columns
```

```python
# function to delete the space at the start of each column name
def delete_space(df):
    df.columns = df.columns.str.lstrip()
    return df
```

```python
df = delete_space(df)
df.columns
```

```python
df.rename(columns = {'cibil_score':'credit_score'},inplace =True)
```

```python
df['credit_grp'] = np.where((df['credit_score']>=300) & (df['credit_score']<=500),"low",
                               np.where((df['credit_score']>500)& (df['credit_score']<600),"mid","high"))
```

```python
df['credit_grp'].value_counts()
```

```python
sns.countplot(x='credit_grp',hue='loan_status',data=df)
plt.show()
```

```python
# Histogram for income
plt.hist(df['income_annum'], bins=10,edgecolor='black')
plt.title('Income Annum')
plt.show()
```

```python
# Histogram for loan amount
plt.hist(df['loan_amount'], bins=10,edgecolor='black')
plt.title('Loan amount')
plt.show()
```

```python
# Boxplot dependents vs loan status classification
sns.boxplot(x='loan_status', y='no_of_dependents', data=df)
plt.title('Dependents vs Loan Status')
plt.show()
```

```python
# Boxplot cibil score vs loan status classification
sns.boxplot(x='loan_status', y='credit_score', data=df)
plt.title('Credit Score vs Loan Status')
plt.show()
```

```python
# Scatterplot Cibil Score vs Loan Amount colored by Loan status
sns.scatterplot(x='credit_score', y='loan_amount', hue='loan_status', data=df)
plt.title('Credit Score vs Loan Amount')
plt.show()
```

```python
loan_term_counts = df['loan_term'].value_counts()
plt.figure(figsize=(8, 8))

plt.pie(loan_term_counts, labels=loan_term_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of loan term ')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.show()
```
